```python

csv.UFO_shape = pd.Categorical(csv.UFO_shape)
csv.UFO_shape = csv.UFO_shape.cat.codes
X = csv.drop(['UFO_shape', 'Unnamed: 0', 'Description', 'Date_time', 'Country_Code','latitude', 'longitude', 'Region', 'Locale', 'date_documented', 'Encounter_Duration','Country'],axis=1)
y = csv['UFO_shape']
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)



X_train.Season = pd.Categorical(X_train.Season)
X_train.Season = X_train.Season.cat.codes

#X_train.Country = pd.Categorical(X_train.Country)
#X_train.Country = X_train.Country.cat.codes

X_test.Season = pd.Categorical(X_test.Season)
X_test.Season = X_test.Season.cat.codes

Dec = DecisionTreeClassifier().fit(X_train, y_train)
c = cross_val_score(Dec, X_train, y_train)
c2 = cross_val_score(Dec, X_test, y_test)
y_pred = Dec.predict(X_test)
print("DecisionTreeClassifier")
print("train")
print(c)
print("%0.2f accuracy with a standard deviation of %0.2f" % (c.mean(), c.std()))
print("test")
print(c2)
print("%0.2f accuracy with a standard deviation of %0.2f" % (c2.mean(), c2.std()))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')

```

# Linear Model

```python
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score,roc_curve, precision_recall_curve, PrecisionRecallDisplay
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)



X_train.Season = pd.Categorical(X_train.Season)
X_train.Season = X_train.Season.cat.codes

#X_train.Country = pd.Categorical(X_train.Country)
#X_train.Country = X_train.Country.cat.codes

X_test.Season = pd.Categorical(X_test.Season)
X_test.Season = X_test.Season.cat.codes

scaler_log = StandardScaler()
X_scaled_tr = scaler_log.fit_transform(X_train)
scaler_log = StandardScaler()
X_scaled_te = scaler_log.fit_transform(X_test)
logreg_model = linear_model.LogisticRegression(max_iter=1500)
logreg_model.fit(X_scaled_tr, y_train)
y_pred = logreg_model.predict(X_scaled_te)
print("Logistic")
print("train")
print(c)
print("%0.2f accuracy with a standard deviation of %0.2f" % (c.mean(), c.std()))
print("test")
print(c2)
print("%0.2f accuracy with a standard deviation of %0.2f" % (c2.mean(), c2.std()))
#sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')

y_proba = logreg_model.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_proba, pos_label=1)
print("Roc AUC:", roc_auc_score(y_test, logreg_model.predict_proba(X_test)[:,1],average='macro'))
plt.plot(fpr, tpr, label="ff")

prec, recall, _ = precision_recall_curve(y_test, y_proba, pos_label=logreg_model.classes_[1])
pr_display = PrecisionRecallDisplay(precision=prec, recall=recall).plot()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score,roc_curve, precision_recall_curve, PrecisionRecallDisplay



X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)



X_train.Season = pd.Categorical(X_train.Season)
X_train.Season = X_train.Season.cat.codes

#X_train.Country = pd.Categorical(X_train.Country)
#X_train.Country = X_train.Country.cat.codes

X_test.Season = pd.Categorical(X_test.Season)
X_test.Season = X_test.Season.cat.codes

RF = RandomForestClassifier().fit(X_train, y_train)
c = cross_val_score(RF, X_train, y_train)
c2 = cross_val_score(RF, X_test, y_test)
y_pred = RF.predict(X_test)
print("RandomForestClassifier")
print("train")
print(c)
print("%0.2f accuracy with a standard deviation of %0.2f" % (c.mean(), c.std()))
print("test")
print(c2)
print("%0.2f accuracy with a standard deviation of %0.2f" % (c2.mean(), c2.std()))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')

#si es desactiva el heatmap que fa la matriu de confusio i es reactiva el de sota es pot veure les corbes

#y_proba = RF.predict_proba(X_test)[:,1]
#fpr, tpr, thresholds = roc_curve(y_test, y_proba, pos_label=1)
#print("Roc AUC:", roc_auc_score(y_test, RF.predict_proba(X_test)[:,1],average='macro'))
#plt.plot(fpr, tpr, label="ff")

#prec, recall, _ = precision_recall_curve(y_test, y_proba, pos_label=RF.classes_[1])
#pr_display = PrecisionRecallDisplay(precision=prec, recall=recall).plot()
```
